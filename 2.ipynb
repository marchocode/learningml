### Cost Function

  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
from common import compute_cost,line_regression, draw_regression,get_train_set
from mpl_toolkits.mplot3d import Axes3D
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730])

w_test = get_train_set(200,300,1)
b_test = get_train_set(1.0,3.2,0.1)
```

```python
length = len(w_test) * len(b_test)
print(length)

xs = np.zeros(length)
ys = np.zeros(length)
zs = np.zeros(length)

index = 0
m = len(x_train)

for w in w_test:
    
    for b in b_test:
       
        cost = compute_cost(w,b,x_train,y_train)
        
        xs[index] = w
        ys[index] = b
        zs[index] = cost
        index += 1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs, ys, zs, s=20)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
idx = np.argmin(zs)
print(xs[idx],ys[idx],zs[idx])

draw_regression(xs[idx],ys[idx],x_train,y_train)
```
