```python
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from common import load_house_data
```

```python
# define scaler
scaler = StandardScaler()
X,y = load_house_data()

# 缩放，并且获得mu 与sigma
X_norm = scaler.fit_transform(X)

print(f"X ptp {np.ptp(X,axis=0)} X shape {X.shape}")
print(f"X_norm ptp {np.ptp(X_norm,axis=0)}")
print(f"scaler mu = {scaler.mean_} and sigma^2 = {scaler.var_} and sigma = {scaler.scale_}")
```

```python
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(X_norm,y)

print(f"model params: w={sgdr.coef_} b={sgdr.intercept_}")
```

```python
x_house = np.array([[1200, 3, 1, 40]])
x_house_norm = scaler.transform(x_house)
prediction = sgdr.predict(x_house_norm)
print(prediction)
```
