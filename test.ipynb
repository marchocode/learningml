### Cost Function

  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 

```python
def compute_cost(x,y,w,b):
    """
    Compute the cost function
    :param x: 
    :param y: 
    :param w: 
    :param b: 
    :return: 
    """
    
    m = x.shape[0]
    cost_sum = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum += cost
    
    total_cost = (1 / (2 * m)) * cost_sum
    
    return total_cost
```
