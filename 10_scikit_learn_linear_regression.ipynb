```python
import numpy as np
from sklearn.linear_model import LinearRegression
from common import load_house_data
```

```python
X,y = load_house_data()

print(X.shape,y.shape)
```

```python
model = LinearRegression()

model.fit(X,y)
print(f"model w={model.coef_} b={model.intercept_}")
```

```python
x_house = np.array([[1200, 3, 1, 40]])

prediction = model.predict(x_house)
print(prediction)
```
