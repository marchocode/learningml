```python

# define a vector
a = np.zeros(5); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")
a = np.zeros((6,)); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")
a = np.random.random_sample(4); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")
```

## Operations on Vectors

```python
a = np.arange(10); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")

print(f"access last element: {a[-1]}")
```

### Slicing

```python
print(a)
c = a[2:7:1]; print("a[2:7:1] =", c)
c = a[2:7:2]; print("a[2:7:2] =", c)
c = a[3:]; print("a[3:]=", c)
c = a[:3]; print("a[:3] =", c)
c = a[:] ; print("a[:] =", c)
```

### Single Vector operations

```python
a = np.array([1,2,3,4]); print(a)
b = -a; print(f"b = -a; b = {b}")
b = np.sum(a); print("b = np.sum(a) =", b)
b = np.mean(a); print("b = np.mean(a) =", b)
b = a ** 2; print("b = a ** 2 = ", b)
```

### Vector element-wise operations


$$
a + b = \sum\limits_{i = 0}^{n-1} a_i + b_i
$$

```python
a = np.array([1,2,3,4]); print(a)
b = np.array([-1,-2,3,4]); print(b)
print(f"a + b = {a + b}")

c = 5*a; print(f"c = 5*a = {c}")
```

```python
def vector_dot(v1, v2):
    
    x = 0
    for i in range(v1.shape[0]):
        x = x + v1[i] * v2[i]
    return x

a = np.array([1,2,3,4]); print(a)
b = np.array([-1,4,3,2]); print(b)

print("vector_dot(a,b) =", vector_dot(a,b))
print("np.dot(a,b) =", np.dot(a,b))
```

## Matrix


```python
a = np.zeros((1,5)); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")
a = np.zeros((2,5)); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")
a = np.random.random_sample((2,2)); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")
```

### Operations on Matrices

```python
a = np.arange(6).reshape(-1, 2); print(f"a = {a} a.shape = {a.shape}, a data type = {a.dtype}")

print(f"access a[2,1] = {a[2,1]}, a[2,1] shape = {a[2,1].shape}, a[2,1] dtype = {a[2,1].dtype}")

print(f"access row a[2] = {a[2]}, a[2] shape = {a[2].shape}, type(a[2]) = {type(a[2])}")
```
