```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from common import load_coffee_data
```

```python
X,y = load_coffee_data()
print(X.shape)
print(y.shape)

print(f"first label max and min is {np.max(X[:,0])} and {np.min(X[:,0])}")
print(f"first label max and min is {np.max(X[:,1])} and {np.min(X[:,1])}")
```

```python
# 归一化
norm_l = tf.keras.layers.Normalization(axis=-1)
norm_l.adapt(X)
X_norm = norm_l(X)

print(f"X_norm first label max and min is {np.max(X_norm[:,0])} and {np.min(X_norm[:,0])}")
print(f"X_norm second label max and min is {np.max(X_norm[:,1])} and {np.min(X_norm[:,1])}")
```

```python
model = Sequential([
    Dense(3, activation='sigmoid', name="L1"),
    Dense(1, activation='sigmoid', name="L2"),
])

model.summary()
```

```python
model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
)
```

```python
Xt = np.tile(X_norm,(1000,1))
Yt= np.tile(y,(1000,1))

model.fit(
    Xt,Yt,            
    epochs=10,
)
```

```python
model.summary()
```

```python
X_test = np.array([
    [200,13.9],  # postive example
    [200,17]])   # negative example

X_testn = norm_l(X_test)
predictions = model.predict(X_testn)
print("predictions = \n", predictions)
```
